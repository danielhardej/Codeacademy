```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def coin_flip_experiment(num_trials):
    prop = []   # proportion of two heads
    flips = []  # number of flips
    two_heads_counter = 0

    # let a coin flip be represented by a random choice between 0 and 1, where 0 is tails and 1 is heads
    flip_one = random.randint(0, 1)
    flip_two = random.randint(0, 1)
    for flip in range(num_trials):
        if flip_one == 1 and flip_two == 1:
            two_heads_counter += 1
        prop.append(two_heads_counter/(flip+1))
        flips.append(flip+1)

    return prop, flips
```

```python
# how many times we run the experiment
num_trials = 10000
prop, flips = coin_flip_experiment(num_trials)


# plot all flips and proportion of two heads
plt.plot(flips, prop, label='Experimental Probability')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Two Heads')

plt.hlines(0.25, 0, num_trials, colors='orange', label='True Probability')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a 3 by 3 grid of subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Flatten the axs array to make it easier to iterate over
axs = axs.flatten()

# Iterate over the subplots and plot the proportion of two heads vs. the number of flips
for i in range(9):
    prop, flips = coin_flip_experiment(num_trials)
    axs[i].plot(flips, prop)
    axs[i].hlines(0.25, 0, num_trials, colors='orange', label='True Probability')
    axs[i].set_title(f"Experiment {i+1}")
    # find difference between experimental and true probability
    diff = abs(prop[-1] - 0.25)
    axs[i].text(0.95, 0.95, f"Diff: {diff:.5f}", transform=axs[i].transAxes, ha='right', va='top')



# set the x and y labels
fig.text(0.5, 0.08, 'Number of Flips', ha='center', va='center')
fig.text(0.08, 0.5, 'Proportion of Two Heads', ha='center', va='center', rotation='vertical')
plt.show()
```
