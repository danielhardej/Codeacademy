16. Congratulations! In this project, you saw how EDA can help with the initial data inspection and cleaning process. This is an important step as it helps to keep your datasets clean and reliable.

    Here are some ways you might extend this project if you'd like:
    - Use `.value_counts()` to more fully explore the values in each column.
    - Investigate other outliers in the data that may be easily overlooked.
    - Instead of changing the `0` values in the five columns to `NaN`, try replacing the values with the median or mean of each column.

````python
# use .value_counts() to more fully explore the values in each column
for col in diabetes_data.columns:
    print(diabetes_data[col].value_counts())
    print('\n')
````

````python
# Investigate other outliers in the data that may be easily overlooked.
# For example, the minimum value for BMI is 0.0. This is impossible for a living person, so this indicates a missing value.
# The same goes for the min value of the BloodPressure column.

# print number of rows with BMI == 0
print("Number of Rows with BMI == 0: ", len(diabetes_data[diabetes_data.BMI == 0]))

# print number of rows with BloodPressure == 0
print("Number of Rows with BloodPressure == 0: ", len(diabetes_data[diabetes_data.BloodPressure == 0]))
````

````python
# Instead of changing the `0` values in the five columns to `NaN`, try replacing the values with the median or mean of each column

# replace 0 values with mean or median values
diabetes_data['Glucose'].fillna(diabetes_data['Glucose'].mean(), inplace=True)
diabetes_data['BloodPressure'].fillna(diabetes_data['BloodPressure'].mean(), inplace=True)
diabetes_data['SkinThickness'].fillna(diabetes_data['SkinThickness'].median(), inplace=True)
diabetes_data['Insulin'].fillna(diabetes_data['Insulin'].median(), inplace=True)
diabetes_data['BMI'].fillna(diabetes_data['BMI'].median(), inplace=True)

# show the new diabetes_data with the missing values filled in
display(diabetes_data.head())

# show the new summary statistics
display(diabetes_data.describe())
````
